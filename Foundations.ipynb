## Functions

```python
import numpy as np
```

```python
print("Python list operations:")
a = [1, 2, 3]
b = [4, 5, 6]
print("a+b:", a+b)
```

```python
try:
    print(a*b)
except TypeError:
    print("a*b has no meaning for Python lists")
```

```python
print("numpy array operations:")
a = np.array([1, 2, 3])
b = np.array([4, 5, 6])
print("a+b:", a+b)
print("a*b:", a*b)
```
